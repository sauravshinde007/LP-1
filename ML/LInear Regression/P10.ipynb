```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

#Step 1: Load dataset
data =  pd.read_csv('Regression/1.01. Simple linear regression.csv')
data.head()
```

```python
# Step 2: Data Preprocessing
# Check for missing values
print(data.isnull().sum())
```

```python
## Step 3: Exploratory Data Analysis (EDA)
# Plotting the data

plt.figure(figsize=(8,6))
sns.scatterplot(x=data['SAT'],y=data['GPA'],data=data)
plt.title('Scatter Plot : SAT Score vs GPA')
plt.xlabel('SAT Score')
plt.ylabel('GPA')
plt.show()
```

```python
#Step 4: Find Correlation
correlation = data.corr()
print(f"Coorelation between SAT SCore and GPA : {correlation.iloc[0,1]}")
```

```python
#Step 5: Build Linear Regression Model

x = data[['SAT']]
y = data['GPA']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
#Step 6: Model evalution
print(f"Intercept : {model.intercept_}")
print(f"Coeffiecient : {model.coef_[0]}")
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)

print("MSE : ",mse)
print("R2 Score : ",r2)
```

```python
#Step 7: Plot the graph
plt.figure(figsize=(8,6))
sns.scatterplot(x=data['SAT'], y=data['GPA'], color='blue')
plt.plot(x_test,y_pred,color='red')
plt.title('Linear Regression : SAT Score vs GPA')
plt.xlabel('SAT Score')
plt.ylabel('GPA')

plt.show()
```
