```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("Regression/diabetes.csv")
data
```

```python
x = data[['Glucose']]
y = data[['Outcome']]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

```

```python
# Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Get the model coefficients
intercept = model.intercept_
coefficient = model.coef_[0]
```

```python
# Predict on the test data
y_pred = model.predict(x_test)
```

```python
# Calculate residual sum of squares (RSS)
rss = ((y_test - y_pred) ** 2).sum()
```

```python
#R2 score
r2 = r2_score(y_test,y_pred)
```

```python
# Output the results
print(f"Intercept (β₀): {intercept}")
print(f"Coefficient (β₁): {coefficient}")
print(f"Residual Sum of Squares (RSS): {rss}")
print(f"Coefficient of Determination (R²): {r2}")
```

```python
#Scatter plot of test data
plt.figure(figsize=(8,6))
sns.scatterplot(x=x_test['Glucose'],y=y_test['Outcome'],label = "Actual Data")

#Plotting regression line
plt.plot(x_test['Glucose'],y_pred,color='red',label='Regression Line')

#Adding labels and title
plt.xlabel('Glucose')
plt.ylabel('Outcome')
plt.title('Linear Regression : Glucose vs Outcome')
plt.legend()


#Show plot
plt.show()

```
