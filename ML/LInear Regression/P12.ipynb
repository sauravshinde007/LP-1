```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


data = pd.read_csv('Regression/advertising.csv')
data.head()
```

```python
print(data.isna().sum())
```

```python
x = data[['TV']]
y = data['Sales']

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
#Model evaluation
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("MSE : ",mse)
print("R2 Score : ",r2)
print(f"Model Intercept : {model.intercept_}")
print(f"Model Coeffieceint : {model.coef_[0]}")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='TV',y='Sales',data = data, color='blue')
plt.plot(x_test,y_pred,color='red') #regression line
plt.title('Linear Regression : TV vs Sales')
plt.xlabel('TV')
plt.ylabel('Sales')

plt.show()
```
